```python
# 讀取訓練資料
df_train = pd.read_csv("Intel_dataset_train.csv",index_col="Date",parse_dates=True)
# 倒數第二格是希望模型預測的結果，因此指定為倒數第二格的修正後收盤價
X_train_set = df_train.drop(labels = ['Adj Close'],axis=1).values
Y_train_set = df_train['Adj Close'].values
print(X_train_set.shape)
# 計算總共有幾筆
X_train_len = len(X_train_set)
print("筆數: ", X_train_len)
```

```python
import numpy as np

# 產生特徵資料和標籤資料，預設look_back是一天，如果有給資料則更多
def create_dataset(ds, look_back=1):
    X_data, Y_data = [],[]
    # 每筆訓練資料有 look_back 天的參考資料，標籤為當天的修正後收盤價
    # 第 x 天價格的訓練資料需要有 x-1, x-2, x-3 ...... 到 x-look_back 天
    for i in range(len(ds)-look_back):
        X_data.append(ds[i:(i+look_back), :])
        Y_data.append(ds[i+look_back, :])
    return np.array(X_data), np.array(Y_data)

def create_dataset_forY(ds, look_back=1):
    X_data, Y_data = [],[]
    # 每筆訓練資料有 look_back 天的參考資料，標籤為當天的修正後收盤價
    # 第 x 天價格的訓練資料需要有 x-1, x-2, x-3 ...... 到 x-look_back 天
    for i in range(len(ds)-look_back):
        X_data.append(ds[i:(i+look_back), :])
        Y_data.append(ds[i+look_back, :])
    return np.array(X_data), np.array(Y_data)


# 這邊設定回看5天
look_back = 5

X_train,X_drop = create_dataset(X_train_set, look_back)
Y_train = Y_train_set[5:]
print("回看天數:", look_back)
print("X_train Shape: ", X_train.shape)
print("Y_train Shape: ", Y_train.shape)
print(X_train_set)
print(X_train.shape)
print(Y_train.shape)
```
